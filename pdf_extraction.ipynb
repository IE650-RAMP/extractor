# PDF to TXT Extraction

```python
import pdfplumber

with pdfplumber.open("downloads/wima_wifo/MK_MMDS_2024_2025_14.06.2024.pdf") as pdf:
    # a_page = pdf.pages[5]
    # print(a_page.chars[0])
    all_text = ""
    for page in pdf.pages:
        all_text += page.extract_text()
    print(all_text)
    with open("output.txt", "w", encoding="utf-8") as text_file:
        text_file.write(all_text)
```
